```python
# Importing metrics from sklearn for evaluation
from sklearn import metrics
```

```python
# Creating confusion matrix 
confusion = metrics.confusion_matrix(Scoredata_y_train_pred_final.Converted,Scoredata_y_train_pred_final.Predicted )
confusion
```

```python
# Check the overall accuracy
metrics.accuracy_score(Scoredata_y_train_pred_final.Converted,Scoredata_y_train_pred_final.Predicted)
```

```python
# Substituting the value of true positive
TP = confusion[1,1]
# Substituting the value of true negatives
TN = confusion[0,0]
# Substituting the value of false positives
FP = confusion[0,1] 
# Substituting the value of false negatives
FN = confusion[1,0]
```

```python
# Calculating the sensitivity
TP/(TP+FN)
```

```python
# Calculating the specificity
TN/(TN+FP)
```

```python
# ROC function
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
fpr, tpr, thresholds = metrics.roc_curve(Scoredata_y_train_pred_final.Converted, Scoredata_y_train_pred_final.Conversion_Prob, drop_intermediate = False )
```

```python
# Call the ROC function
draw_roc(Scoredata_y_train_pred_final.Converted, Scoredata_y_train_pred_final.Conversion_Prob)
```
